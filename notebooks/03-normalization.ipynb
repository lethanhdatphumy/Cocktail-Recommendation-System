```python
import pandas as pd
```

```python
# Read the vectorized csv file
df = pd.read_csv("/Users/admin/PycharmProjects/Cocktail-Recommendation-System/database/vectorized_cocktails.csv")
df.head(5)
```

```python
# Step 1: Create categories table (unique values + ID)
categories = pd.DataFrame({"name": df['category'].unique()})
categories["category_id"] = categories.index + 1

# Step 2: Merge category_id into df
df = df.merge(categories, left_on='category', right_on='name', how='left') \
    .rename(columns={'id': 'category_id'})

df.drop(columns=['name_y'], inplace=True)
df.rename(columns={'name_x': 'name'}, inplace=True)
```

```python
# Reorder the columns of categories table
categories = categories[['category_id', 'name']]
```

```python
# Step 3: Create glass types table (unique values + ID)
glass_types = pd.DataFrame({"name": df['glassType'].unique()})
glass_types["glass_type_id"] = glass_types.index + 1
# Step 4: Merge glass_type_id into df
df = df.merge(glass_types, left_on='glassType', right_on='name', how='left') \
    .rename(columns={'id': 'glass_type_id'})
df.drop(columns=['name_y'], inplace=True)
df.rename(columns={'name_x': 'name'}, inplace=True)
```

```python
# Reorder the columns of glass types table
glass_types = glass_types[['glass_type_id', 'name']]
glass_types.head(5)
```

```python
import pandas as pd
import ast
from itertools import chain

# Step 1: Define safe parser
def safe_literal_eval(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except:
            return []
    elif isinstance(val, list):
        return val
    else:
        return []

# Step 2: Apply parser to all ingredient lists
df['ingredients'] = df['ingredients'].apply(safe_literal_eval)

# Step 3: Flatten and clean all ingredient names
all_ingredients = list(chain.from_iterable(df['ingredients']))
all_ingredients = [ing.strip() for ing in all_ingredients if isinstance(ing, str) and ing.strip()]

# Step 4: Get unique list and create table
unique_ingredients = sorted(set(all_ingredients))
ingredients = pd.DataFrame({
    "ingredients": unique_ingredients
})
ingredients["ingredient_id"] = ingredients.index + 1

# Optional: Reorder for clarity
ingredients = ingredients[['ingredient_id', 'ingredients']]
```

```python
# Reorder the columns of ingredient table
ingredients = ingredients[['ingredient_id','ingredients']]
ingredients.head(5)
```

```python
# Create a mapping from ingredient name to ID
ingredient_map = dict(zip(ingredients['ingredients'], ingredients['ingredient_id']))

# Map ingredient names to IDs for each cocktail
df['ingredient_ids'] = df['ingredients'].apply(lambda lst: [ingredient_map.get(ing) for ing in lst if ing in ingredient_map])
```

```python
import ast

def safe_literal_eval(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)
        except:
            return []
    elif isinstance(val, list):
        return val
    else:
        return []

df['ingredient_ids'] = df['ingredient_ids'].apply(safe_literal_eval)
df['ingredientMeasures'] = df['ingredientMeasures'].apply(safe_literal_eval)
```

```python
df['parsed_empty'] = df['ingredient_ids'].apply(lambda x: len(x) == 0)
print(df['parsed_empty'].value_counts())
```

```python
rows = []
for idx, row in df.iterrows():
    cocktail_id = idx + 1
    ingredients = row['ingredient_ids']
    measures = row['ingredientMeasures']

    for i, ingredient_id in enumerate(ingredients):
        measure = measures[i] if i < len(measures) else None
        rows.append({
            'cocktail_id': cocktail_id,
            'ingredient_id': ingredient_id,
            'measure': measure
        })

cocktail_ingredients = pd.DataFrame(rows)
```
