```python
import pandas as pd
```

```python
# Read the vectorized csv file
df = pd.read_csv("/Users/admin/PycharmProjects/Cocktail-Recommendation-System/database/vectorized_cocktails.csv")
df.head(5)
```

```python
# Step 1: Create categories table (unique values + ID)
categories = pd.DataFrame({"name": df['category'].unique()})
categories["category_id"] = categories.index + 1

# Step 2: Merge category_id into df
df = df.merge(categories, left_on='category', right_on='name', how='left') \
    .rename(columns={'id': 'category_id'})

df.drop(columns=['name_y'], inplace=True)
df.rename(columns={'name_x': 'name'}, inplace=True)
```

```python
# Replace the NaN in alcoholic column by 'unknown'
df['alcoholic'] = df['alcoholic'].fillna('unknown')
alcoholic_types = pd.DataFrame({"alcoholic_type": df['alcoholic'].unique()})
alcoholic_types['alcoholic_id'] = alcoholic_types.index + 1
df = df.merge(alcoholic_types, left_on='alcoholic', right_on='alcoholic_type', how='left')
alcoholic_types
```

```python
# Ingredients Table
ingredients = pd.DataFrame({'ingredients':df['ingredients'].unique()})
ingredients['ingredient_id'] = ingredients.index + 1
ingredients.head(5)
```

```python
# drop the duplicate columns
df = df.loc[:, ~df.columns.duplicated()]
df = df.drop(columns=['alcoholic_type','alcoholic_id'])
df.rename(columns={'category_id': 'id'}, inplace=True)
df.head(5)
```

```python
# Cocktails Tables
cocktails = pd.DataFrame({
    'id': df['id'],
    'name': df['name'],
    'category_id': df['category_id'],
    'glassType': df['glassType'],
    'instructions': df['instructions'],
    'image': df['image'],
    'ingredientMeasures': df['ingredientMeasures'],
    'ingredients': df['ingredients'],
    'alcoholic_id': df['alcoholic_id']
})
```
